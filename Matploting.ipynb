```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5,6])
y=np.array([67,70,85,45,56,90])

plt.subplot(4,2,1)
plt.bar(x,y)


x=np.array([1,2,3,4,5,6])
y=np.array([64,56,89,90,40,98])

plt.subplot(4,2,2)
plt.plot(x,y,marker='o',c='hotpink')


x=np.array(["kannada","English","Hindi","Social"])
y=np.array([122,72,78,84])

plt.subplot(4,2,3)
plt.pie(y,labels=x)

x=np.array([1,2,3,4,5,6])
y=np.array([30,10,40,28,90,56])

plt.subplot(4,2,4)
plt.plot(x,y,marker='o',c='g')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Months":["Jan","Feb","Mar","Apr","May","Jun"],
    "Eloctranics":[25,30,15,23,38,40],
    "Glasaris":[15,20,41,28,45,51],
    "Clothings":[23,12,45,25,41,32]
}

df=pd.DataFrame(data)
print(df)


df.plot(x="Months",marker='o')
plt.title("Show Monthly Sales")
plt.ylabel("Sales")
plt.show()
```
