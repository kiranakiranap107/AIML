```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([20,25,30,42,35])
plt.plot(x,y,marker='o',c='green',ms=8,linestyle='dashed',linewidth=3.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(['Soraba','Tirthalli','Sagara','Hosanagara','Badra','Shikaripura'])
y=np.array([20,15,30,40,55,60])
plt.bar(x,y,color='magenta')

plt.title("Survey of Shivamogga")
plt.xlabel('Taluks')
plt.ylabel('People')


plt.grid(axis='y')
plt.show()
```
