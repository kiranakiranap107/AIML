# Create the Read the 2 CSV file given below and perform the following.

# a) Merge the 2 dataframe student ID .

```python
import pandas as pd
```

```python
data={
    "Std_ID":["S001",'S002','S003','S004'],
    "name":["Anu","ravi","Meena","Kiran"],
    "Dept":['CSE','ECE','CSE','MECH']
}
df=pd.DataFrame(data)
df
```

```python
data1={
    "Std_ID":['S001','S002','S003','S005'],
    "math_marks":[85,90,75,80],
    "science_marks":[78,88,82,79]
}
df1=pd.DataFrame(data1)
df1
```

# a) Merge the 2 dataframe on student ID.

```python
# Inner

df_me=pd.merge(df,df1,on='Std_ID',how='inner')
df_me
```

```python
# Outer

df_me=pd.merge(df,df1,on='Std_ID',how='outer')
df_me
```

# b) add a new arttribute called Total_marks by summing math and science.

```python
df_me['Total_Marks']=df1['math_marks']+df1['science_marks']
df_me['Total_Marks']
df_me

```

# c) add new data object to the final dataframe.

```python
df_me.loc[len(df_me.index)]=['S006','adi','AUTO',90,97,187]
df_me
```
