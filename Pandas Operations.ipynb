# Create a Title for plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y,marker='o')
plt.title('Sports Watch Data')
plt.xlabel('Average pulse')
plt.ylabel('Calories Burnage')
plt.show()
```

# Set fonts properties for Title and Lable

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

f1={'family':'serif','color':'blue','size':20}
f2={'family':'serif','color':'darkred','size':15}

plt.plot(x,y,marker='o')
plt.title('Sports Watch Data',fontdict=f1)
plt.xlabel('Average pulse',fontdict=f2)
plt.ylabel('Calories Burnage',fontdict=f2)
plt.show()
```

# Position of Title

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

f1={'family':'serif','color':'blue','size':20}
f2={'family':'serif','color':'darkred','size':15}

plt.plot(x,y,marker='o')
plt.title('Sports Watch Data',fontdict=f1,loc='left')
plt.xlabel('Average pulse',fontdict=f2)
plt.ylabel('Calories Burnage',fontdict=f2)
plt.show()
```

# Grid Lines

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title('Sports Watch Data')
plt.xlabel('Average pulse')
plt.ylabel('Calories Burnage')

plt.plot(x,y,marker='o')

plt.grid(axis='x')
plt.grid(axis='y')
plt.show()
```
